```python
#SERIES DATATYPE:

'''
About Series??
--> Series is a 1dim datatype which can store Homogeneous data values and its size is immutable. Similar to Numpy Array/Vectors.
    But here the shape can't be changed although the values of the Series can be changed without any errors.
    
    Difference here is Series when compared to Numpy array is it can have indexed axis labels instead of just a number 
    location and can hold any type of data/Python object.
'''
```

```python
#Creating a SERIES???

srs_marks = [75,77,97]
std_names = ['A','B','C']

mrks_srs = pd.Series(data=srs_marks)

print(type(mrks_srs))
print(mrks_srs)
```

```python
mrks_srs1 = pd.Series(data=srs_marks, index = std_names)
print(mrks_srs1)
```

```python
mrks_srs2 = pd.Series(std_names,srs_marks)
print(mrks_srs2)
```

```python
#Operation on Series Data???

srs1 = [75,76,70,65,87]
lbl1 = ['a','b','c','d','e']

srs2 = [80,85,96,65,75]
lbl1 = ['a','b','c','d','e']
```

```python
srs_pd1 = pd.Series(srs1,lbl1)
srs_pd2 = pd.Series(srs2,lbl1)
```

```python
srs_avg = (srs_pd1 + srs_pd2)/2
```

```python
srs_avg
```

```python
srs_sum = srs_pd1 + srs_pd2
srs_sum
```

#DATAFRAMES IN PANDAS???

So Dataframes are the most commonly used data structures to load, read and manipulate the data sets for Machine Learning 
in Python. DataFrames are 2dim(rows,columns) tabular structure in shape and can be created using "pd.DataFrame" notation using Pandas.
There size is mutable that is there shape and size can be changed.

SYNTAX -> pandas.DataFrame( data, index, columns)

```python
#Create an Empty DataFrame??

import pandas as pd

df = pd.DataFrame()
print(df)
```

```python
#Create a DataFrame from Lists??

lst = [['Aman',10],['Bhuvan',12],['Carry',15]]


df = pd.DataFrame(lst,columns=['Name','Age'])
print(df)
```

```python
#Data Frame From Dictionary???

lst = {'Name':['Aman', 'Bhuvan', 'Carry', 'Akash'],'Age':[28,27,29,22]}

df = pd.DataFrame(lst)
print(df)
```
